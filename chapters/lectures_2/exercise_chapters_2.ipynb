# Exercise 2.1 (Graphical Analysis)



**Consider the first-order system:**

$$
\dot{x} = \begin{cases}
-x^5 + 2x^3 - x & \text{if } x \le 1, \\
0 & \text{if } 1 < x \le 2, \\
-x + 2 & \text{if } x > 2,
\end{cases}
$$

Dynamics represented in the figure show multiple equilibria. We analyze stability for $x^* = -1, 0, 1, 2$ and the interval $(1, 2)$.

### **Lời giải chi tiết:**

Dựa vào đồ thị $\dot{x}$ vs $x$ và hàm số đã cho:

1.  **Xét điểm $x^* = -1$:**
    * Lân cận trái ($x < -1$): $\dot{x} > 0 \Rightarrow$ vật tiến về $x^*$.
    * Lân cận phải ($x > -1$): $\dot{x} > 0 \Rightarrow$ vật đi xa khỏi $x^*$.
    $\Rightarrow x^* = -1$ **là điểm không ổn định (Unstable)**.

2.  **Xét điểm $x^* = 0$:**
    * Lân cận trái và phải đều có xu hướng tiến về 0.
    * Để xác định loại ổn định, ta khảo sát đạo hàm $f(x) = -x^5 + 2x^3 - x$:
        $$f'(x) = -5x^4 + 6x^2 - 1$$
    * Tại $x=0$: $f'(0) = -1 < 0$.
    $\Rightarrow x^* = 0$ **là điểm ổn định mũ (Exponentially Stable)**.

3.  **Xét điểm $x^* = 1$:**
    * Lân cận trái: $\dot{x} < 0 \Rightarrow$ vật đi xa khỏi 1.
    $\Rightarrow x^* = 1$ **là điểm không ổn định (Unstable)**.

4.  **Xét điểm $x^* = 2$:**
    * Lân cận trái: $\dot{x} = 0 \Rightarrow$ vật đứng yên.
    * Lân cận phải: $\dot{x} < 0 \Rightarrow$ vật tiến về 2.
    
     $\Rightarrow x^* = 2$ **là điểm ổn định theo Lyapunov (Stable i.s.L.)**.
    
    *(Lưu ý: Không ổn định tiệm cận vì phía trái không hội tụ về điểm cân bằng mà chỉ đứng yên)*.

5.  **Xét các điểm $x^* \in (1, 2)$:**
    * Vật ở bất kỳ vị trí nào thuộc khoảng này đều có vận tốc $\dot{x} = 0$ (đứng yên).
    $\Rightarrow$ **Ổn định theo Lyapunov (Stable i.s.L.)**.

# Exercise 2.2 (Basin of Attraction & Bifurcation)



**System:**
$$
\dot{x} = f(x) = -x^3 + 4x^2 + 11x - 30
$$

### **Lời giải chi tiết:**

**a. & b. Xác định điểm cân bằng và tính ổn định:**

Giải phương trình $f(x) = 0$ để tìm điểm cân bằng:

* **Điểm 1** ($x \approx -3$):
    * Đồ thị đi từ dương sang âm ($\searrow$).
    * $\Rightarrow$ **Ổn định tiệm cận (Asymptotically Stable)**.
    * Miền hấp dẫn: $(-\infty, 2)$ (giả sử điểm giữa mất ổn định là 2).
* **Điểm 2** ($x = 2$):
    * Đồ thị đi từ âm sang dương ($\nearrow$).
    * $\Rightarrow$ **Không ổn định (Unstable)**.
* **Điểm 3** ($x = 5$):
    * Đồ thị đi từ dương sang âm ($\searrow$).
    * $\Rightarrow$ **Ổn định tiệm cận (Asymptotically Stable)**.
    * Miền hấp dẫn: $(2, +\infty)$.

**c. Phân tích Bifurcation với tham số $w$ ($\dot{x} = f(x) + w$):**

Điểm cực tiểu của hàm số $f(x)$ có tọa độ $(-1, -36)$.

1.  **Trường hợp $w \in (0, 36)$**:
    * Đồ thị cắt trục hoành tại 3 điểm.
    * $\Rightarrow$ **3 điểm cân bằng**: Ổn định - Mất ổn định - Ổn định.

2.  **Trường hợp $w = 36$**:
    * Đồ thị tiếp xúc trục hoành tại cực tiểu.
    * $\Rightarrow$ **2 điểm cân bằng**: 1 điểm "bán ổn định" (semistable) và 1 điểm Ổn định tiệm cận.

3.  **Trường hợp $w > 36$**:
    * Đồ thị dịch lên cao, chỉ cắt trục hoành tại 1 điểm nhánh phải.
    * $\Rightarrow$ **1 điểm cân bằng**: Ổn định tiệm cận.

# Exercise 2.3 (Enumerate Unstable Equilibria)



**Given:** Continuous system $\dot{x} = f(x)$ with exactly **three** stable (i.s.L.) equilibrium points. Find $n_{\min}$ and $n_{\max}$ (unstable equilibria).

### **Lời giải chi tiết:**

Hệ có 3 điểm ổn định trên trục hoành (gọi là $S_1, S_2, S_3$).

* **Tìm $n_{\min}$ (Số điểm mất ổn định tối thiểu):**
    * Tại lân cận mỗi điểm ổn định, vector vận tốc hướng về điểm đó ($\rightarrow S \leftarrow$).
    * Xét khoảng giữa $S_1$ và $S_2$: Để dòng chảy đổi chiều từ ($\rightarrow S_1 \leftarrow$) sang ($\rightarrow S_2 \leftarrow$), bắt buộc phải có ít nhất một điểm ở giữa đóng vai trò "đẩy" vật ra ($\leftarrow U \rightarrow$).
    * Tương tự cho khoảng giữa $S_2$ và $S_3$.
    * $\Rightarrow$ **Cần tối thiểu 2 điểm mất ổn định** (1 điểm cho mỗi khoảng).
    * $\Rightarrow n_{\min} = 2$.

* **Tìm $n_{\max}$ (Số điểm mất ổn định tối đa):**
    * Hàm $f(x)$ liên tục có thể dao động tùy ý. Giữa hai điểm ổn định, đồ thị có thể cắt hoặc tiếp xúc với trục hoành vô số lần.
    * Đặc biệt, nếu đồ thị tạo ra một đoạn nằm ngang trùng với trục hoành, ta sẽ có vô số điểm cân bằng.
    * $\Rightarrow n_{\max} = \infty$ (vô cùng).

# Exercise 2.4 (Attractivity vs Stability)



**System:**
$$
\begin{cases}
\dot{x}_1 = x_1(1 - |\mathbf{x}|) + x_2 \frac{x_1 - |\mathbf{x}|}{2|\mathbf{x}|} \\
\dot{x}_2 = x_2(1 - |\mathbf{x}|) - x_1 \frac{x_1 - |\mathbf{x}|}{2|\mathbf{x}|}
\end{cases}
$$
với $|\mathbf{x}| = \sqrt{x_1^2 + x_2^2}$.

### **Lời giải chi tiết:**

Dựa vào biểu đồ pha (Streamplot) và phân tích vector:

1.  **Tại gốc tọa độ (Điểm A):**
    * Các vector dòng chảy hướng ra xa khỏi gốc tọa độ ($r=0$).
    * $\Rightarrow$ **Điểm không ổn định (Unstable)**.

2.  **Tại đường tròn đơn vị $|\mathbf{x}| = 1$ (Vùng B):**
    * Nếu vật ở trong ($|\mathbf{x}| < 1$): Quỹ đạo xoắn ốc hướng ra ngoài, tiến về $r=1$.
    * Nếu vật ở ngoài ($|\mathbf{x}| > 1$): Quỹ đạo xoắn ốc hướng vào trong, tiến về $r=1$.
    * $\Rightarrow$ **Chu trình giới hạn ổn định (Stable Limit Cycle)** hay tập hợp điểm ổn định hấp dẫn.

# Exercise 2.5 (Pendulum with Vibrating Base)



**Equation of Motion:**
$$ml^2 \ddot{\theta} + mgl \sin \theta = ml\omega^2 h \sin(\omega t) \cos \theta + u$$

### **Lời giải chi tiết:**

**a. Xác định động học mong muốn:**
Để vận tốc hội tụ về 1 (gọi $q = \dot{\theta}$), ta chọn:
$$f(q) = -k(q - 1) \quad (\text{với } k > 0)$$
Suy ra: $\ddot{\theta} = -k(\dot{\theta} - 1)$

**b. Luật điều khiển $u$:**
$$u = mgl \sin \theta - ml\omega^2 h \sin(\omega t) \cos \theta - ml^2 k (\dot{\theta} - 1)$$
#### **c. Thực hiện mô phỏng (Implementation)**

Ta thay luật điều khiển $u$ đã tìm được ở câu b vào đoạn code gốc. Luật điều khiển được chia thành hai thành phần để cài đặt:

1.  **Thành phần khử phi tuyến (Cancellation term):**
    Khử ảnh hưởng của trọng lực và lực quán tính do dao động nền:
    $$u_{\text{cancel}} = mgl \sin(\theta) - ml\omega^2 h \sin(\omega t) \cos(\theta)$$

2.  **Thành phần điều khiển mong muốn (Desired dynamics term):**
    Đưa vận tốc góc $\dot{\theta}$ về giá trị 1:
    $$u_{\text{desired}} = -ml^2 k (\dot{\theta} - 1)$$

**Tổng hợp mô-men điều khiển:**
$$\text{torque} = u = u_{\text{cancel}} + u_{\text{desired}}$$

*(Xem đoạn code Python cài đặt chi tiết ở cell bên dưới)*

```python
import numpy as np

def pendulum_torque(m, g, l, theta, theta_dot, t):

    k = 1

    u_cancel = m*g*l*np.sin(theta) - m*l*(omega**2)*h*np.sin(omega*t)*np.cos(theta)

    u_desired = -m*l**2*k*(theta_dot - 1.0)

    torque = u_cancel + u_desired
    return torque
```
